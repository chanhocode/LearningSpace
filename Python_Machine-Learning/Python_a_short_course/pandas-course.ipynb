## Missing Data 
- missing data   row  columns  : dropna()

```python
df = pd.DataFrame({'A':[1,np.nan,3], 'B':[5,np.nan,np.nan], 'C':[1,2,3]})

# missing value    column 

df.dropna(axis=1)
```
