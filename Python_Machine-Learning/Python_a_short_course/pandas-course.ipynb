## Missing Data 처리
- missing data 가 있는 row 혹은 columns을 완전히 삭제: dropna()
- 임의의 data로 대체 : fillna()

```python
df = pd.DataFrame({'A':[1,np.nan,3], 'B':[5,np.nan,np.nan], 'C':[1,2,3]})

# missing value를 포함하고 있는 모든 column 삭제

df.dropna(axis=1)
```

```python
# missing value를 포함하고 있는 모든 row 삭제
df.dropna(axis=0)
```
