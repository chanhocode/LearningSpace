```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 
plt.plot(3,2,'o',c='r')
plt.show()
```

```python
# 
plt.plot([1,2,3],[4,5,1])
```

```python
# title, label 
x = [5,8,10]
y = [12,16,6]

x2=[6,9,11]
y2=[6,15,7]

plt.plot(x, y)
plt.plot(x2, y2)

plt.title('Test Plot')
plt.xlabel('month')
plt.ylabel('price')
plt.show()
```

```python
# bar chart
plt.bar(x,y, label='korea')
plt.bar(x2, y2, label='us')
plt.legend()
```

```python
# scatter plot ()
plt.scatter(x,y)
plt.scatter(x2, y2)
```

```python
# histogram
np.random.seed(0)
x = np.random.randn(1000)
plt.hist(x, bins=20)
plt.show()
```
